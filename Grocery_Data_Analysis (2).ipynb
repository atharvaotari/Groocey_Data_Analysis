```python
Category_name = 'Snacks'  # Change to the item you want to analyze
desired_month = 1  # Change to the desired month (e.g., April)

# Filter data for the specified item
Category_data = df[df['Category'] == Category_name]

# Convert 'Date' column to datetime if it's not already
Category_data['Sales'] = pd.to_datetime(Category_data['Sales'])

# Filter data for the desired month
Category_data_month = Category_data[Category_data['Sales'].dt.month == desired_month]

# Step 3: Plot the scatter graph
plt.scatter(Category_data_month['Sales'], Category_data_month['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profits')
plt.title(f'Profits of {Category_name} in Month {desired_month}')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
Category_counts = df['Category'].value_counts()

# Step 3: Sort the products based on their counts
sorted_products = Category_counts.sort_values(ascending=False)

# Step 4: Select the top 5 products
top_5_products = sorted_products.head(5)

# Step 5: Plot the top 5 products on a graph
plt.figure(figsize=(10, 6))
top_5_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Grocery Products')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
